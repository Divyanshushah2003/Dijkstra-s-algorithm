```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.Graph()

G.add_nodes_from([1, 2, 3, 4, 5])

G.add_weighted_edges_from([(1, 2, 1), (1, 3, 4), (2, 3, 2), 
                           (2, 4, 5), (3, 5, 3), (4, 5, 1)])

source = 1

shortest_paths = nx.single_source_dijkstra_path(G, source)

shortest_path_lengths = nx.single_source_dijkstra_path_length(G, source)

print("Shortest paths from node", source, ":")

for target, path in shortest_paths.items():
    print(f"Path to {target}: {path} with total cost {shortest_path_lengths[target]}")

pos = nx.spring_layout(G)

nx.draw(G, pos, with_labels=True, node_color='lightblue', node_size=500, font_size=10)

nx.draw_networkx_edge_labels(G, pos, edge_labels={(u, v): f'{d["weight"]}' for u, v, d in G.edges(data=True)})

plt.show()
```
